# Data Visuaisation using matplotlib


```python
#import matplotlib
import matplotlib.pyplot as plt
```

## Scatterplot - value in the data set is represented by a dot.

```python
#scatterplot using scatter() method
x = [1,2,3,4,5]
y = [10,15,20,25,30]
plt.scatter(x,y)
plt.show()
```

## Bar plot - value in the data set is represented by bar graphs


```python
#Bar plot using bar() method
plt.bar(x,y)
plt.show()
```

## Pie plot - represents data in the form of pie charts

```python
#Pie chart using pie() method
plt.pie(x)
plt.show()
```

```python
plt.pie(y)
plt.show()
```

## Subplots - to draw multiple plots in one figure

```python
#Subplots using subplot() method
z = [5,9,7,3,4]
#plot 1
plt.subplot(1,2,1)
plt.plot(x,y)

#plot 2
plt.subplot(1,2,2)
plt.plot(x,z)

plt.show()
```
